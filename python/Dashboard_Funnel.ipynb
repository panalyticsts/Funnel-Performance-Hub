```python
import pandas as pd
import numpy as np
print('Librerías cargadas correctamente')
```

```python
# Cargar el archivo (si lo subiste directamente a Colab):
df = pd.read_csv('user_table.csv')

# Si lo tienes en Google Drive, usa esta línea en su lugar:
# df = pd.read_csv('/content/drive/MyDrive/user_table.csv')

print(f'Filas cargadas: {len(df)}')
print(f'Columnas: {list(df.columns)}')
df.head()
```

```python
# Convertir la columna date a formato fecha
df['date'] = pd.to_datetime(df['date'])

# Crear columnas adicionales
df['year']  = df['date'].dt.year
df['month'] = df['date'].dt.month
df['month_name'] = df['date'].dt.strftime('%B')
df['year_month'] = df['date'].dt.strftime('%Y-%m')

# Crear columna de etapa del funnel basada en el user_id
def asignar_etapa(uid):
    uid_str = str(uid)
    if '1' in uid_str: return 'lead'
    elif '2' in uid_str: return 'qualified'
    elif '3' in uid_str: return 'proposal'
    elif '4' in uid_str: return 'closed_won'
    else: return 'lead'

df['stage'] = df['user_id'].apply(asignar_etapa)
df['source'] = df['device']
df['segment'] = df['sex']

print('Transformación completada')
df.head(3)
```

```python
# Guardar el archivo limpio
df.to_csv('funnel_events.csv', index=False)
print('Archivo guardado: funnel_events.csv')
print(f'Total registros: {len(df)}')
```
